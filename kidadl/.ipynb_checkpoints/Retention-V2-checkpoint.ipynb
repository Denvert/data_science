```python
import pandas as pd
import datetime

from os import listdir
from os.path import isfile, join
```

```python
mypath = '/home/skein/kidadl/Retention Bars/'
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
```

```python
files_to_use = list(filter(lambda f: 'Retention' in f and 'csv' in f, onlyfiles))
files_to_use = ['Retention Bars/' + f for f in files_to_use]
files_to_use
```

```python
list_ = []

for file_ in files_to_use:
    df = pd.read_csv(file_,index_col=None, header=0)
    list_.append(df)

data = pd.concat(list_, axis = 0, ignore_index = True)
```

```python
from datetime import datetime

def days_between(d1, d2):
    d1 = datetime.strptime(d1, "%Y%m%d")
    d2 = datetime.strptime(d2, "%Y%m%d")
    return abs((d2 - d1).days)
```

```python
period_days = 7
data['to_stay'] = data.apply(lambda row: 1 if (int(row['Date']) - int(row['date_of_install'])) <= period_days 
                             else 0, axis=1)
#data['to_stay'] = data.apply(lambda row: 1 if days_between(row['date_of_install'], row['Date']) < 90 and 
#                             days_between(row['date_of_install'], row['Date']) > 60
#                             else 0, axis=1)
```
